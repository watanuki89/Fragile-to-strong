```python
import numpy as np
import math
import matplotlib        as mpl
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
mpl.rc('font', family = 'serif', serif = 'Computer Modern Roman', size = 28)
mpl.rc('text', usetex = True)
mpl.rc('legend', fontsize = 19)
mpl.rc('xtick', labelsize = 22)
mpl.rc('ytick', labelsize = 22)
def VFT(T,D0,A,T0):
	return (D0 * np.exp(-A/(T-T0)))
def Arrh(T,DA,E):
	return (DA * np.exp(-E/(8.617e-5*T)))
D = np.loadtxt('Diff.dat', usecols=[0],'' unpack=True)
T = np.array([1900, 1800, 1700, 1600, 1500, 1400, 1300, 1200, 1100, 1000, 990, 980, 970, 950, 880, 850, 840, 820, 800, 790, 780, 770, 760, 750, 740, 730, 720, 710, 700, 690, 680, 670, 660])
popt, pcov = curve_fit(VFT, T[10:22], D[10:22])
popt3, pcov3 = curve_fit(Arrh, T[28:-1], D[28:-1])
fig = plt.figure()
ax = fig.add_axes([0.18, 0.17, 0.8, 0.78])
ax.set_xlim(0.8, 1.6)
ax.plot(1000.0/T[7:],1.0/D[7:],'ko')
ax.plot(1000.0/T[7:31], 1.0/VFT(T, *popt)[7:31], 'r',label='VFT')
ax.plot(1000.0/T[18:], 1.0/Arrh(T, *popt3)[18:], 'b--',label='ARR')
ax.set_yscale('log')
ax.set_xlabel('$1000/T(K)$')
ax.set_ylabel('$1/D(m^2/s)$')
ax.set_facecolor('white')
ax.legend(loc = 'best', fancybox = True, shadow = True)
fig.savefig("Diff_fit.pdf",facecolor='white', edgecolor='white')
plt.show()
fig.clf()
```
