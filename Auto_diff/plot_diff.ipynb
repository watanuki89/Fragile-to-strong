```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family = 'serif', serif = 'Computer Modern Roman', size = 18)
plt.rc('text', usetex = True)
plt.rc('xtick', labelsize=16) 
plt.rc('ytick', labelsize=16)
D_Zr = np.loadtxt('Diff_Zr.dat', unpack=True,usecols=[0])
D_Cu = np.loadtxt('Diff_Cu.dat', unpack=True,usecols=[0])
D = np.loadtxt('Diff.dat', unpack=True,usecols=[0])
T = np.loadtxt('temp.dat', unpack=True,usecols=[0])
Dr = 1.0/D
```

```python
fig = plt.figure()
#fig.suptitle(r'\phantom{a} Driving force', size = 24)
ax = fig.add_axes([0.18, 0.17, 0.8, 0.78])
ax.set_xlim(0.5, 1.6)
ax.set_ylim(1e-13, 2e-8)
#ax.set_yticks(np.arange(1e-13, 1e-7, 1e-1))
#ax.set_xticks(np.arange(0.4,1.5,0.2))
ax.plot(1000.0/T,D_Zr,'yo', label = 'Zr')
ax.plot(1000.0/T,D_Cu,'mo', label = 'Cu')
#ax.plot(T,D,'ko', label = 'All')
ax.set_facecolor('white')
ax.tick_params(axis = 'both', which = 'major', labelsize = 20)
plt.xlabel('$1000/T$',fontsize=20)
plt.ylabel('$D(m^2/s)$',fontsize=20)
ax.set_yscale('log')
#ax.set_xscale('log')
#plt.grid('on')
ax.legend(loc = 'best', fancybox = True, shadow = True)
#ax.legend(loc=3,prop={'size':15})
fig.savefig("Diff.pdf",facecolor='white', edgecolor='white')
plt.show()
fig.clf()
```

```python
fig = plt.figure()
plt.plot(1000.0/T,Dr,'ko')
plt.yscale('log')
fig.savefig("Diff_inv.pdf",facecolor='white', edgecolor='white')
#plt.axis([1.2, 1.6, 0,0.4e13])
plt.show()
```
