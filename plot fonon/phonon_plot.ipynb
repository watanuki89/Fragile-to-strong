```python
from scipy.fftpack import fft, fftfreq, ifft
import matplotlib.pyplot as plt
import numpy             as np
import math
import matplotlib        as mpl
mpl.rc('font', family = 'serif', serif = 'Computer Modern Roman', size = 28)
mpl.rc('text', usetex = True)
mpl.rc('legend', fontsize = 19)
mpl.rc('xtick', labelsize = 22)
mpl.rc('ytick', labelsize = 22)
T = [1800, 1500, 1200, 1000, 800, 750, 700]
gr = [[]] * len(T)
w = np.loadtxt('omega.dat' , usecols=[0], unpack=True) 
i = 0
fig = plt.figure()
ax = fig.add_axes([0.18, 0.17, 0.8, 0.78])
for n in T:
    gr[i] = np.loadtxt('gr_%i.dat' %(n), usecols=[0], unpack=True)
    ax.plot(w[4:],gr[i][4:], label = '%i' %(n))
    i += 1
ax.set_xlim(0, 50)
ax.set_ylim(0.0, 0.5)
ax.set_xlabel('$\omega$ (THz)')
ax.set_ylabel('g($\omega$)/$\omega^2$')
ax.set_facecolor('white')
ax.legend(loc = 'best', fancybox = True, shadow = True)
fig.savefig("phonon.pdf",facecolor='white', edgecolor='white')
plt.show()
fig.clf()
```
